```python

df['day'] =  pd.to_datetime(df['day'], format='%Y-%m-%d %H:%M:%S')
import datetime
current_date = max(df['day']) + datetime.timedelta(days=1)
df_customers = df.groupby(['customer_id']).agg(
    {'day': lambda x: (current_date- x.max()).days,
     'order_id':'count',
     'total_sales':'sum'
     }
)
df_customers['customer_id']=df['customer_id'].unique()
```

```python
df_customers.rename(columns={'day': 'Recency'}, inplace=True)
df_customers.rename(columns={'order_id': 'Frequency'}, inplace=True)
df_customers.rename(columns={'total_sales': 'Monetary'}, inplace=True)
df_customers = df_customers.rename_axis('customer_id', axis=1)
df_customers.columns
```

```python
df_customers['recency(25)'] = pd.qcut(df_customers['Recency'], q=[0, 0.2, 0.4, 0.6, 0.8, 1], labels=False, duplicates='drop')
df_customers['Frequency(25)'] = pd.qcut(df_customers['Frequency'], q=[0, 0.2, 0.4, 0.6, 0.8, 1], labels=False, duplicates='drop')
df_customers['Monetary(25)'] = pd.qcut(df_customers['Monetary'], q=[0, 0.2, 0.4, 0.6, 0.8, 1], labels=False, duplicates='drop')
```

The Recency chart demonstrates a relatively even distribution, with a large number of customers making recent purchases within the last 10 days. While this pattern is typical in purchasing behavior models, it may indicate some challenges, such as customers making only a few purchases before the products no longer suit their needs. In the Monetary chart, the majority of customers appear to spend less than 50,000 units of currency.

```python
def calculate_recency_score(recency):
    if recency == 4:
        return 5
    elif  recency == 3:
        return 4
    elif  recency == 2:
        return 3
    elif recency == 1:
        return 2
    elif recency == 0:
        return 1
df_customers['recency_score'] = df_customers['recency(25)'].apply(calculate_recency_score)

print(df_customers)
```

```python
def calculate_Frequency_score(Frequency):
    if Frequency ==4:
        return 5
    elif  Frequency ==3:
        return 4
    elif Frequency == 2:
        return 3
    elif  Frequency == 1:
        return 2
    if Frequency ==0:
        return 1
df_customers['Frequency_score'] = df_customers['Frequency(25)'].apply(calculate_Frequency_score)

print(df_customers)
```

```python
def calculate_Monetary_score(Monetary):
    if Monetary ==4:
        return 5
    elif Monetary == 3:
        return 4
    elif Monetary == 2:
        return 3
    elif Monetary == 1:
        return 2
    if Monetary ==0:
        return 1
df_customers['Monetary_score'] = df_customers['Monetary(25)'].apply(calculate_Monetary_score)

print(df_customers)
```

```python
df_customers['recency_score'] = df_customers['recency_score'].astype(str)
df_customers['Frequency_score'] = df_customers['Frequency_score'].astype(str)
df_customers['Monetary_score'] = df_customers['Monetary_score'].astype(str)
df_customers['RFM_SEGMENT'] = df_customers['recency_score'].str.cat([df_customers['Frequency_score'], df_customers['Monetary_score']], sep='')

print(df_customers)
```

```python
df_customers.rename(columns={'RFM_SEGMENT': 'RFM_SEGMENT_Score'}, inplace=True)
```

```python
def map_RFM_segment(RFM_SEGMENT):
    A = ['555', '554', '544', '545', '454', '455', '445']
    B = ['543', '444', '435', '355', '354', '345', '344', '335']
    C = ['553', '551', '552', '541', '542', '533', '532', '531', '452', '451', '442', '441', '431', '453', '433', '432', '423', '353', '352', '351', '342', '341', '333', '323']
    D = ['512', '511', '422', '421', '412', '411', '311']
    E = ['525', '524', '523', '522', '521', '515', '514', '513', '425', '424', '413', '414', '415', '315', '314', '313']
    F = ['535', '534', '443', '434', '343', '334', '325', '324']
    G = ['155', '154', '144', '214', '215', '115', '114', '113']
    I = ['331', '321', '312', '221', '213']
    J = ['255', '254', '245', '244', '253', '252', '243', '242', '235', '234', '225', '224', '153', '152', '145', '143', '142', '135', '134', '133', '132', '125', '124']
    K = ['332', '322', '231', '241', '251', '233', '232', '223', '222', '132', '123', '122', '212', '211']
    L = ['111', '112', '121', '131', '141', '151']

    if RFM_SEGMENT in A:
        return 'Champion'
    elif RFM_SEGMENT in B:
        return 'Loyal Customer'
    elif RFM_SEGMENT in C:
        return 'Potential Loyalist'
    elif RFM_SEGMENT in D:
        return 'New Customer'
    elif RFM_SEGMENT in E:
        return 'Promising'
    elif RFM_SEGMENT in F:
        return 'Need Attention'
    elif RFM_SEGMENT in G:
        return 'Cannot Lose Them'
    elif RFM_SEGMENT in I:
        return 'About to Sleep'
    elif RFM_SEGMENT in J:
        return 'At Risk'
    elif RFM_SEGMENT in K:
        return 'Hibernating'
    elif RFM_SEGMENT in L:
        return 'Lost'
df_customers['RFM_Segment'] = df_customers['RFM_SEGMENT_Score'].map(map_RFM_segment)

print(df_customers)
```

```python
df['RFM_Segment_Score'] = df['customer_id'].map(df_customers.set_index('customer_id')['RFM_SEGMENT_Score'])
df['RFM_Segment'] = df['customer_id'].map(df_customers.set_index('customer_id')['RFM_Segment'])
df['Rencency'] = df['customer_id'].map(df_customers.set_index('customer_id')['Recency'])
df['Frequency'] = df['customer_id'].map(df_customers.set_index('customer_id')['Frequency'])
df['Monetary'] = df['customer_id'].map(df_customers.set_index('customer_id')['Monetary'])

# In DataFrame sau khi thm ct mi
print(df)
```

```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import plotly.graph_objects as go
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
print(df_customers)
```

```python
X = df_customers.loc[:, ['Recency', 'Frequency']].values
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X)
    sse.append(kmeans.inertia_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o',)
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

After drawing the Elbow to select K, we see that K = 2 is the most optimal for this case because the following Ks have begun to gradually decrease. To be more certain, we should use the silhouette score index.

```python
from yellowbrick.cluster import SilhouetteVisualizer
for n_clusters in [2, 3, 4]:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X)
    silhouette_visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    silhouette_visualizer.fit(X)
    silhouette_visualizer.show()
```

The silhouette score is a metric used to assess the quality of clustering, ranging from -1 to 1:

+ A value close to 1 indicates that data points are correctly assigned to clusters.
+ A value close to 0 indicates that data points lie close to the boundary between clusters.
+ A value close to -1 indicates that data points may have been assigned to the wrong cluster.

In this case, we use the silhouette score to plot for k=2,3,4:

+ For k=2, the silhouette score is approximately 0.7, which is the highest among the three scores.
+ We can also observe that the cluster widths are more evenly distributed compared to the other clusters, indicating that the number of points in each cluster is fairly balanced.

```python
kmeans = KMeans(n_clusters=2)
labels = kmeans.fit_predict(X)
df_customers['Labels'] = labels
print(df_customers)
```

## S dng ct recency v monetary

```python
from scipy.stats import zscore
df_customers['Recency_scaled'] = zscore(df_customers['Recency'])
df_customers['Monetary_scaled'] = zscore(df_customers['Monetary'])
print(df_customers[['Recency_scaled', 'Monetary_scaled']])
```

```python
X1 = df_customers.loc[:, ['Recency_scaled', 'Monetary_scaled']].values
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
    kmeans.fit(X1)
    sse.append(kmeans.inertia_)
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), sse, marker='o',)
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.xticks(range(1, 11))
plt.grid(True)
plt.show()
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
for n_clusters in [3, 4, 5]:
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(X1)
    silhouette_visualizer = SilhouetteVisualizer(kmeans, colors='yellowbrick')
    silhouette_visualizer.fit(X1)
    silhouette_visualizer.show()
```

```python
kmeans = KMeans(n_clusters=3)
labels = kmeans.fit_predict(X1)
df_customers['Labels 2'] = labels
print(df_customers)
```
